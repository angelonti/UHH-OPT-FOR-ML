## Task 2: Affine invariance

Prove the following lemma.

**Lemma**

Let $f\colon \mathbb{R}^d \to \mathbb{R}$ be twice differentiable, $A \in \mathbb{R}^{d \times d}$ an invertible matrix, $b \in \mathbb{R}^d$.
Let $g\colon \mathbb{R}^d \to \mathbb{R}^d$ be the (bijective) function
$$
g(x) := Ax + b.
$$
Finally, for a twice differentiable function $h\colon \mathbb{R}^d \to \mathbb{R}$, let $N_h\colon \mathbb{R}^d \to \mathbb{R}^d$ denote the Newton step for $h$, i.e.
$$
N_h(x) := x - \nabla^2 h(x)^{-1} \nabla h(x),
$$
whenever this is definde. Then
$$
N_{f \circ g} := g^{-1} \circ N_f \circ g.
$$

**Proof**

Calculate the Gradient and Hessian of $h(x)$ which is $f(g(x))$ or $f\circ g$.
$$
h(x) = f(g(x)) = f(Ax + b) \\
\nabla h(x) = A^T \nabla f(Ax + b) = A^T \nabla f(g(x)) \\
\nabla^2 h(x) = A^T \nabla^2 f(Ax + b) A = A^T \nabla^2 f(g(x)) A
$$
Plug in the gradient and hessian into the newton step of $h(x)$.
$$
\begin{align}
N_{f \circ g}(x) &= x - (A^T \nabla^2 f(g(x)) A)^{-1} (A^T \nabla f(g(x))) \\
&= x - A^{-1} (\nabla^2 f(g(x)))^{-1} \nabla f(g(x))
\end{align}
$$

Now we have to show that $N_{f \circ g}(x) = g^{-1} \circ N_f \circ g(x)$.
$$
\begin{align}
N_f &= x - (\nabla^2 f(x))^{-1} \nabla f(x) \\
g^{-1}(y) &= A^{-1} (y - b) \\
\text{Consider } y &= N_f \circ g(x) \\
&= Ax+b - (\nabla^2 f(Ax+b))^{-1} \nabla f(Ax+b) \\
\end{align}
$$
Then we have to show that $g^{-1}(y) = N_{f \circ g}(x)$.
$$
\begin{align}
g^{-1}(y) &= g^{-1}\circ N_f\circ g\\ 
&= A^{-1} (Ax+b - (\nabla^2 f(Ax+b))^{-1} \nabla f(Ax+b) -b) \\
&= x - A^{-1} (\nabla^2 f(Ax+b))^{-1} \nabla f(Ax+b) \\
&= x - A^{-1} (\nabla^2 f(g(x)))^{-1} \nabla f(g(x)) \\
&= N_{f \circ g}(x) \ \square
\end{align}
$$

## Task 3: Non-quadratic convergence example

Let $\delta > 0$ be a real number. Find an example of a convex function $f \colon \mathbb{R} \to \mathbb{R}$ such that Newton's method satisfies
$$
|x_{t+1} - x^\star| \geq (1 - \delta) |x_t + x^\star|
$$
for all $x_t \neq x^\star$. For this function, Newton's method will not have a local quadratic convergence phase.

Consider the function $f(x) = x^4$, which is a convex function. The first and second derivative are:
$$f'(x) = 4x^3$$
$$f''(x) = 12x^2$$

Setting the first derivative to zero, we get the minimum is $x^\star = 0$

The Newton step is:
$$
x_{t+1} = x_t - \frac{f'(x_t)}{f''(x_t)} = x_t - \frac{4x_t^3}{12x_t^2} = x_t - \frac{1}{3}x_t = \frac{2}{3}x_t
$$

Plugging in the minimum $x^\star = 0$, and the Newton step into the inequality:

$$
|x_{t+1} - x^\star| \geq (1 - \delta) |x_t + x^\star|
$$

We get:

$$
\frac{2}{3}\left|x_t\right| \geq (1 - \delta) |x_t|
$$

This implies that $\delta \geq \frac{1}{3}$. $\square$

For each step, the error (difference between the current guess and the actual square root) is halved or more. That is, if we let $e_t = x_t - \sqrt{R}$, then $|e_{t+1}| \leq \frac{1}{2} |e_t|$.

Assuming our initial guess was $x_0 = R$, then $|e_0| = |R - \sqrt{R}| \leq R$.

Therefore, the number of iterations needed to achieve $|e_t| < \frac{1}{2}$ is at most the number of times we can halve $R$ until we get a number less than $\frac{1}{2}$. This number is $\log_2 R - 1 = O(\log R)$.

## Task 5: Positive definite bound

Prove that $\|Mx\|_2 \leq L\|x\|_2$ for all positive semidefinite $M$ with $L\cdot \mathbb{I} \succeq M$.

Since $M$ is positive semidefinite, and $L$ is a positive scalar, $L \cdot \mathbb{I} \succeq M$ implies:

$$
x^TMx \leq x^T(L \cdot \mathbb{I})x
$$

Now, let's consider the 2-norm of $Mx$:

$$
\|Mx\|_2^2 = (Mx)^T(Mx) = x^TM^TMx
$$

Since $L \cdot \mathbb{I} \succeq M$, we can replace $M^TM$ with $(L \cdot \mathbb{I})^T(L \cdot \mathbb{I})$ in the expression for $\|Mx\|_2^2$:

$$
\|Mx\|_2^2 = x^TM^TMx \leq x^T(L \cdot \mathbb{I})^T(L \cdot \mathbb{I})x = x^TL^TLx = \|Lx\|_2^2
$$

Now, taking the square root of both sides and using the fact that the 2-norm is non-negative:

$$
\|Mx\|_2 \leq \|Lx\|_2
$$

Now, since $L$ is a positive scalar, we can factor it out of the norm:

$$
\|Mx\|_2 \leq L\|x\|_2  \ \square
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
def contour_map(f, xb=(-1,1), yb=(-1,1), ax=None):
    if ax is None:
        ax = plt
    (nx, ny) = (45, 45)
    x = np.linspace(*xb, nx)
    y = np.linspace(*yb, ny)
    xv, yv = np.meshgrid(x, y)
    X = np.block([ [xv.reshape(1, -1)], [yv.reshape(1, -1)] ]).T
    zv = np.fromiter((f(x) for x in X), dtype=np.double)
    zv = zv.reshape(nx,ny)
    ax.contour(xv, yv, zv, 15)

def surface_plot(f, xb=(-1,1), yb=(-1,1)):
    (nx, ny) = (45, 45)
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    x = np.linspace(*xb, nx)
    y = np.linspace(*yb, ny)
    xv, yv = np.meshgrid(x, y)
    X = np.block([ [xv.reshape(1, -1)], [yv.reshape(1, -1)] ]).T
    zv = np.fromiter((f(x) for x in X), dtype=np.double)
    zv = zv.reshape(nx,ny)
    ax.plot_surface(xv, yv, zv, cmap=cm.coolwarm)
    return fig, ax
```

```python
def backtracking_line_search(x, d, f, g, alpha=0.3, beta=0.8):
    step_size = 1.
    while f(x + step_size * d) > f(x) + alpha * step_size * g(x).dot(d):
        step_size *= beta
    return step_size

def backtracking_line_search_nag(x, d, f, g, alpha=1., beta=0.8):
    t = 1
    while f(x + t*d) > f(x) + alpha * t * np.dot(g(x), d) + t/2*np.dot(d, d):
        t *= beta
    return t
```

```python
def gradient_descent_path(x0, f, g, max_iter=100):
    xs = [x0]
    for _ in range(max_iter):
        step = backtracking_line_search(x0, -g(x0), f, g)
        x0 = x0 - step * g(x0)
        xs.append(x0)
    return xs

def heavy_ball_path(x0, f, g, max_iter=100):
    xs = [x0]
    x_old = x0
    for i in range(1, max_iter+1):
        d = -g(x0) + (i-1)/(i+1) *  (x0-x_old)
        x_old = x0
        t = backtracking_line_search_nag(x0, d, f, g)
        x0 = x0 + t * d
        xs.append(x0)
    return xs

def nag_path(x0, f, g, max_iter=100):
    xs = [x0]
    x_old = x0
    for i in range(1, max_iter+1, 1):
        y = x0 + (i-1)/(i+2) *  (x0-x_old)
        x_old = x0
        t = backtracking_line_search_nag(y, -g(y), f, g)
        x0 = y - t * g(y)
        xs.append(x0)
    return np.array(xs)
```

```python
EXA = np.array([[30., 15],[-20, 25]])/20

EXAMPLES = [
    (lambda x: x.T@EXA@x + 1, lambda x: (EXA+EXA.T)@x, lambda x: (EXA+EXA.T), 0.5 * np.ones(2), (-1.,1.), (-1.,1.)),
    (lambda x: (x[0]**2 + 30 * x[1]**2 + 4 * x[0]), lambda x: np.array([2 * x[0] + 4, 60 * x[1]]), lambda x: np.array([[2, 0],[0, 60]]), np.array([2.,3.]), (-2.5,2.5), (-1.5,3.5)),
    (lambda x: np.linalg.norm(np.sin(x*3))**2, lambda x: 6 * np.sin(x*3) * np.cos(x*3), lambda x: 18 * np.diag(2 * np.cos(3*x)**2 - 1), np.array([.2,.15]), (-.5, .5), (-.5,.5))
]

def run_examples_newtons_method():
    for (f, g, h, x0, xb, yb) in EXAMPLES:
        xs2 = newtons_method_path(x0, f, g, h)
        xs2 = np.array(xs2)
        
        xs = gradient_descent_path(x0, f, g)
        xs = np.array(xs)

        contour_map(f, xb=xb, yb=yb)
        plt.plot(xs2[:,0], xs2[:,1], '.--k', label='newtons method')
        plt.plot(xs[:,0], xs[:,1], '.--', color='gray', alpha=.5, label='gradient descent')
        plt.legend()
        plt.show()
```

## Task 6: Newton's method

Implement Newton's method using the above provided backtracking line search.
* `x0` is the initial point.
* `f` is the function you are trying to minimize.
* `g` is the gradient of `f`.
* `h` is the hessian of `f`.

Function `newtons_method_path` should return a list of vectors on the path to the minimum.

```python
def newtons_method_path(x0, f, g, h, max_iter=100):
    xs = [x0]
    for _ in range(max_iter):
        d = np.linalg.solve(h(x0), -g(x0))
        t = backtracking_line_search(x0, d, f, g)
        x0 = x0 + t * d
        xs.append(x0)
    return xs
```
